   echo "Welcome to the Moving Target Guessing Game!"
   ```



2. **Start Prompt**: It waits for the player to press Enter to start the game using the `read` command.

   ```bash
   read -p "Press Enter to start..."
   ```


3. **Initialize Target Position**: It initializes the position of the target by generating a random number between 1 and 100.

   ```bash
   target_position=$((1 + RANDOM % 100))



4. **Display Target**: The `display_target` function clears the terminal screen, displays a message indicating the game's purpose, and creates a horizontal line to represent the target area. The target position is marked with an 'X'.

   ```bash
   function display_target {
       clear  # Clear the terminal screen
       echo "Moving Target Aim Trainer - Hit the target area between 1 and 100"
       echo ""
       
       # Display the target area as a horizontal line
       target_line=""
       for ((i = 1; i <= 100; i++)); do
           if [[ $i -eq $target_position ]]; then
               target_line+="X"
           else
               target_line+="-"
           fi
       done
   
       echo "$target_line"
       echo ""
   }
   ```


5. **Check Guess**: The `check_guess` function prompts the player to enter their guess for the target's position. It then checks if the guess matches the current target position and provides feedback to the player.

   ```bash
   function check_guess {
       read -p "Enter your guess (1-100): " guess
   
       if [[ $guess -eq $target_position ]]; then
           echo "Congratulations! You hit the target at position $target_position."
       else
           echo "Missed. The target was at position $target_position."
       fi
   }
